```python
import pandas as pd
from string import ascii_lowercase
from statistics import mean
import numpy as np
from multiprocess import Pool # https://stackoverflow.com/questions/41385708/multiprocessing-example-giving-attributeerror
import dill
import functools
import operator

dill.settings['recurse'] = True # https://github.com/uqfoundation/pathos/issues/129
```

```python
# Read in and concatenate solution words and valid guesses
solutions = pd.read_table('solutions.txt', header=None)
solutions['solution'] = True
guesses = pd.read_table('guesses.txt', header=None)
guesses['solution'] = False
df = pd.concat([solutions, guesses]).reset_index(drop=True)
df.columns = ['word', 'solution']
df
```

```python
# Split words into letters and join back onto df
letters = df.word.str.split('', expand=True).loc[:, 1:5]
df = df.join(letters)
df
```

```python
# Count the number of occurrences of each letter in each word
for letter in ascii_lowercase:
    df[letter] = df.word.str.count(letter)

# Set the full word as index, for convenience
df = df.set_index('word')

# Store a filtered DF with only valid solutions, for use later
df_solution = df[df.solution]

df
```

```python
def count_possibilities(solution, guess):
    '''Returns the number of possible solutions for a given solution/guess pair.
    
    Parameters
    ----------    
    solution : string
        The solution word
    
    guess : string
        The guess word
    
    Note: This function expects DataFrames named 'df' and 'df_solution'
    to be available in the global namespace. Sue me.
    
    Returns
    -------
    int
        The number of possible solutions based on the information obtained
        for this solution and guess
    '''
    sub_df = df_solution.copy()
    
    # Select subsections of the DF based on whether each position is a match
    for i in range(1, 6):
        if sub_df.loc[solution, i] == df.loc[guess, i]:
            sub_df = sub_df[sub_df[i] == guess[i - 1]]
        else:
            sub_df = sub_df[sub_df[i] != guess[i - 1]]

    # Select subsections of the DF based on comparing letter counts
    for letter in guess:
        if sub_df.loc[solution, letter] >= df.loc[guess, letter]:
            # If the solution had at least as many occurrences of the letter as the guess did,
            # we learn only that there must be at least that many occurrences of the letter
            sub_df = sub_df[sub_df[letter] >= df.loc[guess, letter]]
        else:
            # If the guess had more occurrences of the letter than the solution did,
            # we learn the exact number of occurrences of the letter in the solution
            sub_df = sub_df[sub_df[letter] == sub_df.loc[solution, letter]]
            
    return len(sub_df)
```

```python
def count_possibilities_functools(solution, guess):
    '''Returns the number of possible solutions for a given solution/guess pair.
    
    Parameters
    ----------    
    solution : string
        The solution word
    
    guess : string
        The guess word
    
    Note: This function expects DataFrames named 'df' and 'df_solution'
    to be available in the global namespace. Sue me.
    
    Returns
    -------
    int
        The number of possible solutions based on the information obtained
        for this solution and guess
    '''
    masks = []
    
    # Select subsections of the DF based on whether each position is a match
    for i in range(1, 6):
        if df_solution.loc[solution, i] == df.loc[guess, i]:
            masks.append(df_solution[i] == guess[i - 1])
        else:
            masks.append(df_solution[i] != guess[i - 1])

    # Select subsections of the DF based on comparing letter counts
    for letter in guess:
        if df_solution.loc[solution, letter] >= df.loc[guess, letter]:
            # If the solution had at least as many occurrences of the letter as the guess did,
            # we learn only that there must be at least that many occurrences of the letter
            masks.append(df_solution[letter] >= df.loc[guess, letter])
        else:
            # If the guess had more occurrences of the letter than the solution did,
            # we learn the exact number of occurrences of the letter in the solution
            masks.append(df_solution[letter] == df_solution.loc[solution, letter])
            
    return sum(functools.reduce(operator.and_, masks)) # https://stackoverflow.com/questions/31144126/python-bitwise-and-on-multiple-numbers-faster-way-than-iterative-bitwise-operat
```

```python
count_possibilities('aware', 'error') == count_possibilities_functools('aware', 'error')
```

```python
def count_possibilities_df(guesses):
    out_df = pd.DataFrame()
    
    for guess in guesses:
        print(f'Processing {guess}...')
        for solution in df_solution.index:
            # Guesses on the rows, solutions on the columns
            out_df.loc[guess, solution] = count_possibilities_functools(solution, guess)
    
    return out_df.astype(int)
```

```python
def get_average_possibilities(guess):
    print(f'Processing {guess}...')
    return mean([count_possibilities(solution, guess) for solution in df_solution.index])
```

```python
def get_all_average_possibilities(df):
    df['avg_num_possibilities'] = df['word'].apply(get_average_possibilities)
    return df
```

```python
def parallelize_dataframe(df, func, n_cores=4):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
```

```python
out_df = parallelize_dataframe(df.index, count_possibilities_df, 8)
out_df.to_csv('out_matrix.csv')
out_df
```
